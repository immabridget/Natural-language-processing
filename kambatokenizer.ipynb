```python
import re
regex = r'(?=[?!,.:;])'
#s = "abracadabra abraca|dabara | abra cadabra abra ca dabra abra ca dabra abra"
f = open("kamba.txt","r",encoding="utf8")
lines = f.readlines()

f1 = open("kamba_dt.txt","w+",encoding="utf8")

for line in lines:
    l = re.sub(regex, ' \g<0>', line)
    f1.write(l)
    print(l)
    print('-'*30)
```
